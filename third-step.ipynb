```python
# Membuat Plot
import numpy as np
import matplotlib.pyplot as plt

#Membuat Data
def sinusGenerator(amplitudo, frekuensi,waktu, theta):
    t = np.arange(0, waktu,0.1)
    y = amplitudo * np.sin(2*frekuensi*t + np.deg2rad(theta))
    return t,y

#Membuat plot
t1,y1 = sinusGenerator(1,1,4,0)
t2,y2 = sinusGenerator(1,1,4,-90)
t3,y3 = sinusGenerator(1,1,4,-30)
t4,y4 = sinusGenerator(1,1,4,20)
dataPlot1 = plt.plot(t1,y1, label="sin(0)")
dataPlot2 = plt.plot(t2,y2, label="sin(-90)")
dataPlot3 = plt.plot(t3,y3, label="sin(-30)")
dataPlot4 = plt.plot(t4,y4, label="sin(20)")
# plt.legend(loc="upper right") #Didalam ktaok
plt.legend(loc="upper right", bbox_to_anchor=(0.6,-0.1)) #Diluar ktaok


#Setting Label dan Title
judul = 'Grafik Sinusoidal\n'
rumus = r'$\mathcal{Y} = A\sin(2\pi f t + \theta)$'
plt.title(judul+rumus)

plt.xlabel('waktu(m/s)')
plt.ylabel('magnituda(cm)')
#Setting Properties
plt.setp(dataPlot1, color='r', linestyle='-', linewidth=2)
plt.setp(dataPlot2, color='g', linestyle='--', linewidth=2)
plt.setp(dataPlot3, color='b', linestyle='-.', linewidth=2)
plt.setp(dataPlot4, color='black', linestyle=':', linewidth=2)


#Menampilkan Plot
plt.show()
```

```python
#STUDI KASUS
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Melihat Data
mhs = pd.read_excel("mahasiswa.xlsx")

#Mengolah Data

saintek = [
    "Teknik Informatika",
    "Sistem Informasi",
    "Ilmu Komputer",
    "Pendidikan Dokter",
    "Teknik Mesin",
    "Desain Komunikasi Visual",
    "Teknik Kimia",
    "Fisika",
    "Bioteknologi",
    "Keperawatan",
    "Arsitektur",
    "Kesehatan Masyarakat",
    "Teknik Lingkungan",
    "Teknik Industri",
    "Teknik Komputer",
    "Teknik Elektro",
    "Teknik Otomotif",
    "Seni Rupa",
    "Biologi",
    "Teknik Geologi",
    "Matematika",
    "Ilmu Lingkungan",
    "Teknologi Pangan",
    "Teknik Biomedis",
    "Teknik Geodesi",
    "Kedokteran",
    "Teknik Komputer",
    "Teknik Pertambangan",
    "Ilmu Kebumian"
]
soshum = [
    "Akuntansi",
    "Manajemen",
    "Hukum",
    "Psikologi",
    "Ilmu Politik",
    "Ekonomi",
    "Ilmu Komunikasi",
    "Pendidikan Bahasa Inggris",
    "Desain Produk",
    "Ilmu Perekonomian",
    "Ilmu Hukum",
    "Teknologi Informasi",
    "Ilmu Komunikasi"
]
sumSaintek = 1
sumSoshum = 1
#Menganalisis relasi jurusan dan bekerja
for i in range(len(mhs)):
    if mhs["Jurusan"][i] in saintek:
        if mhs["Status Pekerjaan"][i] == "Bekerja":
            sumSaintek += 1
    elif mhs["Jurusan"][i] in soshum:
        if mhs["Status Pekerjaan"][i] == "Bekerja":
            sumSoshum += 1

saintekBekerja =  ( sumSaintek/ len(mhs)) * 100
soshumBekerja =  ( sumSoshum/ len(mhs)) * 100



labels = ['Saintek', 'Soshum']
sizes = [saintekBekerja, soshumBekerja]
colors = ['blue', 'green']
fig, data = plt.subplots()
data.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90, wedgeprops={'edgecolor': 'black'})
data.set_title('Persentase Mahasiswa yang Bekerja Berdasarkan Jurusan', fontsize=14, pad=20)
if sumSaintek > sumSoshum:
    data.text(1.2, 0, f"Berdasarkan Analisis yang diberikan ternyata presentase \n mahasiswa dengan jurusan soshum lebih sedikit dapat \n pekerjaan")
else:
    data.text(1.2, 0, f"Berdasarkan Analisis yang diberikan ternyata presentase \n mahasiswa dengan jurusan saintek lebih sedikit dapat \n pekerjaan")
# Inisialisasi penghitung untuk jabatan
ketua_internasional = 0
ketua_nasional = 0
wakil_nasional = 0
anggota_nasional = 0
wakil_internasional = 0
anggota_internasional = 0

# Menganalisis jabatan berdasarkan keterangan Nasional/Internasional
for i in range(len(mhs)):
    # Memeriksa jabatan Ketua dan apakah magang Internasional
    if mhs["Jabatan"][i] == "Ketua":
        if mhs["Keterangan Nasional/Internasional"][i] == "Internasional":
            ketua_internasional += 1
        elif mhs["Keterangan Nasional/Internasional"][i] == "Nasional":
            ketua_nasional += 1
    # Memeriksa jabatan Wakil dan apakah magang Nasional atau Internasional
    elif mhs["Jabatan"][i] == "Wakil":
        if mhs["Keterangan Nasional/Internasional"][i] == "Nasional":
            wakil_nasional += 1
        elif mhs["Keterangan Nasional/Internasional"][i] == "Internasional":
            wakil_internasional += 1

    # Memeriksa jabatan Anggota dan apakah magang Nasional atau Internasional
    elif mhs["Jabatan"][i] == "Anggota":
        if mhs["Keterangan Nasional/Internasional"][i] == "Nasional":
            anggota_nasional += 1
        elif mhs["Keterangan Nasional/Internasional"][i] == "Internasional":
            anggota_internasional += 1

# Data hasil analisis
labels = ["Ketua Magang Internasional","Ketua  Magang  Nasional", "Wakil  Magang  Nasional", "Wakil Magang  Internasional", "Anggota Magang Nasional", "Anggota Magang Internasional"]
values = [ketua_internasional, ketua_nasional, wakil_nasional, wakil_internasional, anggota_nasional, anggota_internasional]

# Membuat histogram
fig, ax = plt.subplots()
ax.bar(labels, values, color=['blue', "pink", 'green', 'orange', 'red', 'purple'], edgecolor='black')

# Menambahkan judul dan label
ax.set_title("Jumlah Jabatan yang diterima Magang (Nasional/Internasional)", fontsize=14)
ax.set_ylabel("Jumlah", fontsize=12)
ax.set_xticklabels(labels, rotation=45, ha='right')

# Menambahkan grid untuk memperjelas
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Menampilkan histogram
plt.show()
```
