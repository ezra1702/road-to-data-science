```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file Excel
data = pd.read_excel("pkkmb.xlsx")

# Menampilkan isi data
print("Data PKKMB:")
print(data)

# Menampilkan informasi data
print("\nInformasi Data:")
print(data.info())

# Menampilkan statistik deskriptif
print("\nStatistik Deskriptif:")
print(data.describe())

# Menambahkan kolom Status Kelulusan
def cek_kelulusan(row):
    # Menentukan kelulusan jika semua nilai > 0
    if row["Teknik Komputer"] > 0 and row["Teknik Informatika"] > 0 and row["Teknologi Informasi"] > 0 and row["Sistem Informasi"] > 0:
        return "Lulus"
    else:
        return "Tidak Lulus"

# Mengganti nama kolom untuk pemahaman yang lebih baik
data = data.rename(columns={
    "Tekkom": "Teknik Komputer",
    "TIF": "Teknik Informatika",
    "TI": "Teknologi Informasi",
    "SI": "Sistem Informasi"
})

# Tambahkan kolom baru berdasarkan logika kelulusan
data["Status Kelulusan"] = data.apply(cek_kelulusan, axis=1)

# Menampilkan data dengan kolom Status Kelulusan
print("\nData dengan Status Kelulusan:")
print(data)

# Menyimpan data ke file Excel baru
data.to_excel("hasil_kelulusan_pkkmb.xlsx", index=False)

# Visualisasi: Diagram Lingkaran Kelulusan
kelulusan_counts = data["Status Kelulusan"].value_counts()
kelulusan_counts.plot(kind="pie", autopct="%1.1f%%", labels=kelulusan_counts.index, colors=["green", "red"])
plt.title("Distribusi Kelulusan PKKMB")
plt.ylabel("")  # Menghilangkan label Y pada pie chart
plt.show()
```
