```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data_aktivitas_mingguan = {
    'Hari': ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'],
    'Langkah': [4000, 7500, 6000, 8000, 5000, 10000, 3000],
    'Kalori Terbakar': [200, 350, 300, 400, 250, 500, 150],
    'Jam Tidur': [6, 7, 8, 6, 5, 9, 10],
    'Produktivitas (%)': [70, 80, 75, 85, 65, 90, 50],
    'Kategori Aktivitas': ['Sedang', 'Tinggi', 'Sedang', 'Tinggi', 'Rendah', 'Sangat Tinggi', 'Rendah']
}
df_data_aktivitas_mingguan = pd.DataFrame(data_aktivitas_mingguan)
#Statistik Deskriptif untuk Kategori
#Mean per Kategori diperlihatkan menggunakan visual
langkah = df_data_aktivitas_mingguan['Langkah'].mean()
kalori = df_data_aktivitas_mingguan['Kalori Terbakar'].mean()
jam = df_data_aktivitas_mingguan['Jam Tidur'].mean()
produktivitas = df_data_aktivitas_mingguan['Produktivitas (%)'].mean()
plt.bar(['Langkah', 'Kalori Terbakar', 'Jam Tidur', 'Produktivitas (%)'], [langkah, kalori, jam, produktivitas], color=['blue', 'green', 'red', 'yellow'])
plt.xlabel('Macam - Macam Kategori')
plt.ylabel('Mean (rata - rata tiap kategori)')
plt.show()

#Analisis Hari Tertinggi dan Terendah
tertinggi_kalori = df_data_aktivitas_mingguan['Kalori Terbakar'].idxmax()
terendah_kalori = df_data_aktivitas_mingguan['Produktivitas (%)'].idxmax()
hari_teringgi = df_data_aktivitas_mingguan.loc[tertinggi_kalori, 'Hari']
hari_terendah = df_data_aktivitas_mingguan.loc[terendah_kalori, 'Hari']
print(f"Hari dengan jumlah kalori terbakar tertinggi adalah {hari_teringgi}")
print(f"Hari dengan produktivitas tertinggi adalah {hari_terendah}")

#Visualisasi Performa Mingguan
plt.bar(df_data_aktivitas_mingguan['Hari'], df_data_aktivitas_mingguan['Langkah'], color=['blue', 'green', 'red', 'yellow', 'orange', 'purple', 'pink'])
plt.xlabel('Hari')
plt.ylabel('Langkah')
plt.title('Performa Mingguan')
plt.show()

plt.plot(df_data_aktivitas_mingguan['Hari'], df_data_aktivitas_mingguan['Kalori Terbakar'], marker='o', linestyle='--', color='red')
plt.xlabel('Hari')
plt.ylabel('Kalori Terbakar')
plt.show()

plt.scatter(df_data_aktivitas_mingguan['Langkah'], df_data_aktivitas_mingguan['Kalori Terbakar'], color='green')
plt.xlabel('Langkah')
plt.ylabel('Kalori Terbakar')
plt.show()

# Membuat histogram dengan garis KDE
plt.figure(figsize=(8, 6))
sns.histplot(df_data_aktivitas_mingguan['Produktivitas (%)'], kde=True, color='blue')
plt.xlabel('Langkah')
plt.ylabel('Jumlah')
plt.title('Histogram Langkah')
plt.show()

#Filter data untuk hari dengan Produktivitas (%) di bawah rata-rata.
df_data_aktivitas_mingguan[df_data_aktivitas_mingguan['Produktivitas (%)'] < df_data_aktivitas_mingguan['Produktivitas (%)'].mean()]


print(df_data_aktivitas_mingguan)


# Menambahkan kolom Rasio Langkah-Kalori
df_data_aktivitas_mingguan['Rasio Langkah-Kalori'] = df_data_aktivitas_mingguan['Langkah'] / df_data_aktivitas_mingguan['Kalori Terbakar']

# Mencari hari dengan rasio tertinggi dan terendah
rasio_tertinggi_idx = df_data_aktivitas_mingguan['Rasio Langkah-Kalori'].idxmax()
rasio_terendah_idx = df_data_aktivitas_mingguan['Rasio Langkah-Kalori'].idxmin()

hari_tertinggi_rasio = df_data_aktivitas_mingguan.loc[rasio_tertinggi_idx, 'Hari']
hari_terendah_rasio = df_data_aktivitas_mingguan.loc[rasio_terendah_idx, 'Hari']

# Menampilkan hasil
print("Data dengan Rasio Langkah-Kalori:")
print(df_data_aktivitas_mingguan)

print(f"\nHari dengan rasio Langkah-Kalori tertinggi adalah {hari_tertinggi_rasio}")
print(f"Hari dengan rasio Langkah-Kalori terendah adalah {hari_terendah_rasio}")
```
